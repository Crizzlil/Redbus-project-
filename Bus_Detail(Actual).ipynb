# Kerala Buses

```python
#Importing Data
dK = pd.read_csv('df_K.csv')
dK
```

```python
driver_K = webdriver.Chrome()
#State = []
Bus_Name_K = []
Bus_Type_K = []
Departure_Time_K = []
Arraival_Time_K = []
Travel_Duration_K = []
Ratings_K = []
Price_K = []
Seats_Available_K = []
Route_Name = []
Route_Links = []

for i,r in dK.iterrows():
    link = r["Route_link"]
    Route = r["Route_name"] 

    driver_K.get(link)
    time.sleep(1)

    elements = driver_K.find_elements(By.XPATH, f"//a[contains(@href, '{link}')]")
    for element in elements:
        element.click()
        time.sleep(3)
    try:
        clicks = driver_K.find_element(By.XPATH, "//div[@class='button']")
        clicks.click()
    except:
        continue
    time.sleep(1)

    scrolling = True
    while scrolling:
        old_page_source = driver_K.page_source

        ActionChains(driver_K).send_keys(Keys.PAGE_DOWN).perform()

        time.sleep(1)

        new_page_source = driver_K.page_source

        if new_page_source == old_page_source:
            scrolling = False
    # Extracting bus details
    bus_name = driver_K.find_elements(By.XPATH, '//div[@class="travels lh-24 f-bold d-color"]')
    bus_type = driver_K.find_elements(By.XPATH, '//div[@class="bus-type f-12 m-top-16 l-color evBus"]')
    start_time = driver_K.find_elements(By.XPATH, '//div[@class="dp-time f-19 d-color f-bold"]')
    end_time = driver_K.find_elements(By.XPATH, '//div[@class="bp-time f-19 d-color disp-Inline"]')
    total_duration = driver_K.find_elements(By.XPATH, '//div[@class="dur l-color lh-24"]')

    try:
        rating = driver_K.find_elements(By.XPATH,"//div[contains(@class, 'rating')]/span")
    except:
        continue
    price = driver_K.find_elements(By.XPATH, '//div[@class="fare d-block"]')
    seats = driver_K.find_elements(By.XPATH, '//div[@class="seat-left m-top-30"]')

    for bus in bus_name:
        Bus_Name_K.append(bus.text)
        Route_Links.append(link)
        Route_Name.append(Route)
    for bus_type_elem in bus_type:
        Bus_Type_K.append(bus_type_elem.text)
    for start_time_elem in start_time:
        Departure_Time_K.append(start_time_elem.text)
    for end_time_elem in end_time:
        Arraival_Time_K.append(end_time_elem.text)
    for total_duration_elem in total_duration:
        Travel_Duration_K.append(total_duration_elem.text)
    for ratings in rating:
        Ratings_K.append(ratings.text)
    for price_elem in price:
        Price_K.append(price_elem.text)
    for seat_elem in seats:
        Seats_Available_K.append(seat_elem.text)

driver_K.quit()

print('Done')

```

```python
max_length = max(
    len(Bus_Name_K), len(Bus_Type_K), len(Departure_Time_K),
    len(Arraival_Time_K), len(Travel_Duration_K), len(Ratings_K),
    len(Price_K), len(Seats_Available_K), len(Route_Name), len(Route_Links)
)

# Extend each list to the maximum length with None (or other placeholder values)
def extend_list(lst, length):
    return lst + [None] * (length - len(lst))

Bus_Name_K = extend_list(Bus_Name_K, max_length)
Bus_Type_K = extend_list(Bus_Type_K, max_length)
Departure_Time_K = extend_list(Departure_Time_K, max_length)
Arraival_Time_K = extend_list(Arraival_Time_K, max_length)
Travel_Duration_K = extend_list(Travel_Duration_K, max_length)
Ratings_K = extend_list(Ratings_K, max_length)
Price_K = extend_list(Price_K, max_length)
Seats_Available_K = extend_list(Seats_Available_K, max_length)
Route_Name = extend_list(Route_Name, max_length)
Route_Links = extend_list(Route_Links, max_length)
```

```python


Data = {
    'Bus_Name': Bus_Name_K,
    'Bus_Type': Bus_Type_K,
    'Departure_Time': Departure_Time_K,
    'Arraival_Time': Arraival_Time_K,
    'Travel_Duration': Travel_Duration_K,
    'Ratings': Ratings_K,
    'Price': Price_K,
    'Seats_Available': Seats_Available_K,
    'Route_Name': Route_Name,
    'Route_Links': Route_Links
}

df_K_bus = pd.DataFrame(Data)

# Correct path formatting
path = r"C:\Users\HP\OneDrive\Desktop\Data Science MDT23\Project\Redbus\Bus_details\df_K_bus.csv"

# Save DataFrame to CSV
df_K_bus.to_csv(path, index=False)
```
